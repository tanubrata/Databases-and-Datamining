```python
biogrid = pd.read_csv("AthBIOGRID.txt", sep="\t")
biogrid.head()
```

```python
biomart = pd.read_csv("AthBiomart.txt", sep="\t")
biomart.head()
```

```python
nextgenraw = pd.read_csv("NextGenRaw.txt", sep="\t")
nextgenraw.head()
```

```python
gridgenesA = biogrid[["Systematic Name Interactor A"]]
gridgeneA = gridgenesA.drop_duplicates()
gridgeneA.columns = ["DataName"]
gridgeneA.count()
```

```python
gridgenesB = biogrid[["Systematic Name Interactor B"]]
gridgeneB = gridgenesB.drop_duplicates()
gridgeneB.columns = ["DataName"]
gridgeneB.count()
```

```python
gridgenes = gridgeneA.append(gridgeneB, ignore_index=True)
gridgenes.count() #all genes from biogrid.txt
```

```python
biomartgenes = biomart[["Gene stable ID"]]
biomartgene = biomartgenes.drop_duplicates()
biomartgene.columns = ["DataName"]
biomartgene.count() #all genes from biomart.txt
```

```python
nextgenrawgenes = nextgenraw[["GENE"]]
nextgenrawgene = nextgenrawgenes.drop_duplicates()
nextgenrawgene.columns = ["DataName"]
nextgenrawgene.count() #all genes from nextgenraw.txt
```

```python
genes1 = gridgenes.append(biomartgene, ignore_index=True)
genes2 = genes1.append(nextgenrawgene, ignore_index=True)
genes3 = genes2.drop_duplicates()
#genes3.count() #contains all gene ids from 3 txt files-28661
genes3.insert(1, 'DTID_DataType', 1)
genes3.count() #genes df
```

```python
goterms = biomart[["GO term accession"]]
goterm = goterms.drop_duplicates()
goterm.columns = ["DataName"]
goterm.dropna(subset=['DataName'], inplace=True)
goterm.insert(1, 'DTID_DataType', 2)
goterm #go term df
```

```python
table_exp = {'DataName' : ['Control1', 'Control2', 'Nitrate1', 'Nitrate2']}
experiment = pd.DataFrame(table_exp)
experiment.insert(1, "DTID_DataType", 3)
experiment #experiment names df
```

```python
df1 = genes3.append(goterm, ignore_index=True)
df1.count() #genes + go term
```

```python
df2 = df1.append(experiment, ignore_index=True)
df2.count() #df for data table
```

```python
df2.insert(0, 'DID', range(1, 1+len(df2)))
df2
```

```python
con = sqlite3.connect("td2201_midterm.sqlite")
c = con.cursor()
```

```python
#Data table created here
c.execute("""
            CREATE TABLE Data(DID INT PRIMARY KEY NOT NULL,
            DataName TEXT, DTID_DataType INT,
            FOREIGN KEY(DTID_DataType) REFERENCES DataType(DTID))
""")
```

```python
#number of tables in my db
pd.read_sql("""
            SELECT * FROM sqlite_master
            WHERE type='table';
""", con)
```

```python
df2.to_sql("Data", con, index=False, if_exists="append")
```

```python
pd.read_sql("""
        select * from Data
""", con)
```
